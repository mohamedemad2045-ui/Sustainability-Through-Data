```python
#Data Type Conversion
df['date'] = pd.to_datetime(df['date'])
print(df['date'].dtypes)
```

```python

print(df['year'].unique())
print(df['month'].unique())
```

```python
# Feature Engineering
# ---------------------------------------------------
# 1. Temporal Feature Engineering (Date Features)
# ---------------------------------------------------
df['quarter'] = df['date'].dt.quarter
df['month_name'] = df['date'].dt.strftime('%B')
df['day'] = df['date'].dt.day
df['day_of_week'] = df['date'].dt.dayofweek + 1  
df['day_name'] = df['date'].dt.strftime('%A')
df['week_of_year'] = df['date'].dt.isocalendar().week

# Feature: Season of the year (Updated to use 'Autumn')
def get_season(month):
    if month in [12, 1, 2]: return 'Winter'
    elif month in [3, 4, 5]: return 'Spring'
    elif month in [6, 7, 8]: return 'Summer'
    else: return 'Autumn'

df['season'] = df['date'].dt.month.apply(get_season)


df['is_weekend'] = df['day_of_week'].isin([6, 7])


# ---------------------------------------------------
# 2. Geographical Feature Engineering (Location Features)
# ---------------------------------------------------
asia_countries = ['Japan', 'China', 'India', 'Vietnam', 'Thailand', 'Indonesia', 'Malaysia', 
                  'Res_Philippines', 'Singapore', 'Bangladesh', 'South Korea', 'Kazakhstan', 
                  'UAE', 'Qatar', 'Saudi Arabia', 'Israel', 'Iraq', 'Turkey', 'Pakistan']
north_america = ['United States', 'Canada', 'Mexico']
south_america = ['Brazil', 'Argentina', 'Chile', 'Peru', 'Colombia']
europe = ['Germany', 'France', 'United Kingdom', 'Italy', 'Spain', 'Poland', 'Netherlands', 
          'Belgium', 'Sweden', 'Switzerland', 'Austria', 'Portugal', 'Denmark', 'Finland', 
          'Ireland', 'Czech Republic', 'Hungary', 'Romania', 'Greece']
oceania = ['Australia', 'New Zealand']
africa = ['South Africa', 'Nigeria', ]

def get_continent(country):
    if country in asia_countries: return 'Asia'
    elif country in north_america: return 'North America'
    elif country in south_america: return 'South America'
    elif country in europe: return 'Europe'
    elif country in oceania: return 'Oceania'
    elif country in africa: return 'Africa'
    else: return 'Middle East'

df['continent'] = df['country'].apply(get_continent)


# ---------------------------------------------------
# 3. Advanced Numerical & Interaction Features (Your Features)
# ---------------------------------------------------
# Feature: Response Efficiency (Relates response time to event severity)
df['response_efficiency'] = df['response_time_hours'] / (df['severity'] + 1)

# Feature: Aid to Loss Ratio (Handles division by zero safely if economic impact is 0)
df['aid_to_loss_ratio'] = np.where(
    df['economic_impact_million_usd'] > 0,
    (df['international_aid_million_usd'] / df['economic_impact_million_usd']) * 100,
    0
)

# Feature: Fatality Rate (Handles division by zero safely if total casualties is 0)
df['fatality_rate'] = np.where(
    df['total_casualties'] > 0,
    (df['deaths'] / df['total_casualties']) * 100,
    0
)


# ---------------------------------------------------
# 4. Event Type Feature Engineering (Categorical Binning)
# ---------------------------------------------------

severity_conditions = [
    (df['severity'] <= 2),
    (df['severity'] <= 4),
    (df['severity'] <= 6)
]
severity_choices = ['Low', 'Medium', 'High']
df['severity_category'] = np.select(severity_conditions, severity_choices, default='Severe')

# Feature: Event Family grouping
def get_event_family(event):
    if event in ['Hurricane', 'Tornado', 'Cyclone']: return 'Wind Storm'
    elif event in ['Flood', 'Tsunami']: return 'Water Related'
    elif event in ['Earthquake', 'Volcanic Eruption', 'Landslide']: return 'Geological'
    elif event in ['Heatwave', 'Cold Wave', 'Drought']: return 'Temperature Extreme'
    elif event in ['Wildfire']: return 'Fire'
    else: return 'Other'

df['event_family'] = df['event_type'].apply(get_event_family)


# ---------------------------------------------------
# 5. Impact Feature Engineering (Binning & Categorization)
# ---------------------------------------------------
# Feature: Death Severity
death_conditions = [
    (df['deaths'] == 0),
    (df['deaths'] <= 5),
    (df['deaths'] <= 20)
]
death_choices = ['No Deaths', 'Low Casualties', 'Medium Casualties']
df['death_severity'] = np.select(death_conditions, death_choices, default='High Casualties')

# Feature: Economic Impact Level
econ_conditions = [
    (df['economic_impact_million_usd'] == 0),
    (df['economic_impact_million_usd'] < 1),
    (df['economic_impact_million_usd'] < 10),
    (df['economic_impact_million_usd'] < 100)
]
econ_choices = ['No impact', 'Minor', 'Moderate', 'Major']
df['economic_impact_level'] = np.select(econ_conditions, econ_choices, default='Catastrophic')

# Feature: Response Time Category
response_conditions = [
    (df['response_time_hours'] <= 6),
    (df['response_time_hours'] <= 24)
]
response_choices = ['Fast', 'Normal']
df['response_time_category'] = np.select(response_conditions, response_choices, default='Slow')

# Feature: Infrastructure Damage Level
infra_conditions = [
    (df['infrastructure_damage_score'] == 0),
    (df['infrastructure_damage_score'] < 10),
    (df['infrastructure_damage_score'] < 30),
    (df['infrastructure_damage_score'] < 50)
]
infra_choices = ['No damage', 'Low', 'Medium', 'High']
df['infrastructure_damage_level'] = np.select(infra_conditions, infra_choices, default='Severe')

# Feature: International Aid Required Flag (Boolean)
df['aid_required'] = df['international_aid_million_usd'] > 0

# Preview the newly engineered dataset
df.head()
```

```python
# outlier

numeric_columns = [
    'severity', 'duration_days', 'affected_population', 
    'deaths', 'injuries', 'economic_impact_million_usd','infrastructure_damage_score', 
    'response_time_hours', 'international_aid_million_usd', 
    'total_casualties', 'impact_per_capita'
]


for col in numeric_columns:
    
    if col in df.columns:
        
        
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        
        outliers_count = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
        
        print(f"column: {col}")
        print(f"Outliers count: {outliers_count}")
        print(f"Normal from {lower_bound:.2f} to {upper_bound:.2f}")
        print("-" * 50)

        plt.figure(figsize=(8, 4))
        sns.boxplot(x=df[col], color='skyblue')
        plt.title(f'Boxplot for {col} (Outliers: {outliers_count})')
        plt.xlabel(col)
        plt.grid(axis='x', linestyle='--', alpha=0.7)
        plt.show()

print(
    "\nOutliers were retained because extreme climate events "
    "represent real-world catastrophic scenarios."
)
```

```python
# ---------------------------------------------------
# Most Dangerous Event Types
# ---------------------------------------------------
dangerous_events = df_final.groupby(['event_type', 'event_family']).agg(
    occurrence_count=('event_id', 'count'),
    total_deaths=('deaths', 'sum'),
    total_injuries=('injuries', 'sum'),
    avg_severity=('severity', 'mean'),
    total_impact_million_USD=('economic_impact_million_usd', 'sum')
).reset_index().sort_values(by='total_deaths', ascending=False)
dangerous_events
```

```python
# ---------------------------------------------------
# Most Dangerous Event Types by Continent (Top 20)
# ---------------------------------------------------
dangerous_events_by_continent = df_final.groupby(['continent', 'event_type']).agg(
    occurrence_count=('event_id', 'count'),
    total_deaths=('deaths', 'sum'),
    avg_deaths_per_event=('deaths', 'mean'),
    total_impact_million_USD=('economic_impact_million_usd', 'sum'),
    avg_severity=('severity', 'mean')
).reset_index().sort_values(by=['total_deaths', 'total_impact_million_USD'], ascending=[False, False]).head(20)
dangerous_events_by_continent
```

```python
# ---------------------------------------------------
# Top 10 Most Affected Countries (Economically)
# ---------------------------------------------------
top_10_countries_econ = df_final.groupby(['country', 'continent']).agg(
    event_count=('event_id', 'count'),
    total_deaths=('deaths', 'sum'),
    total_injuries=('injuries', 'sum'),
    total_affected_population=('affected_population', 'sum'),
    total_economic_impact_million_USD=('economic_impact_million_usd', 'sum'),
    avg_response_time_hr=('response_time_hours', 'mean'),
    avg_severity=('severity', 'mean')
).reset_index().sort_values(by='total_economic_impact_million_USD', ascending=False).head(10)
top_10_countries_econ
```

```python
# ---------------------------------------------------
# Response Time Analysis by Event Type
# ---------------------------------------------------
response_by_event_type = df_final.groupby('event_type').agg(
    event_count=('event_id', 'count'),
    avg_response_hours=('response_time_hours', 'mean'),
    fastest_response_hours=('response_time_hours', 'min'),
    slowest_response_hours=('response_time_hours', 'max'),
    avg_damage_score=('infrastructure_damage_score', 'mean'),
    avg_deaths=('deaths', 'mean')
).reset_index().sort_values(by='avg_response_hours')
response_by_event_type
```

```python
# ---------------------------------------------------
# Impact Severity by Continent and Season
# ---------------------------------------------------
continent_season_impact = df_final.groupby(['continent', 'season']).agg(
    event_count=('event_id', 'count'),
    total_deaths=('deaths', 'sum'),
    total_impact_million_USD=('economic_impact_million_usd', 'sum'),
    avg_severity=('severity', 'mean'),
    avg_duration_days=('duration_days', 'mean')
).reset_index().sort_values(by=['continent', 'season'])
continent_season_impact
```

```python
# ---------------------------------------------------
# Correlation: Response Time vs Impact
# ---------------------------------------------------
# Recreating the conditional response categories 
response_conditions = [
    (df_final['response_time_hours'] <= 6),
    (df_final['response_time_hours'] <= 24)
]
response_choices = ['Fast (<6 h)', 'Normal (6-24 h)']
df_final['response_category'] = np.select(response_conditions, response_choices, default='Slow (>24 h)')

response_impact_correlation = df_final.groupby('response_category').agg(
    event_count=('event_id', 'count'),
    avg_deaths=('deaths', 'mean'),
    avg_economic_impact_million_USD=('economic_impact_million_usd', 'mean'),
    avg_damage_score=('infrastructure_damage_score', 'mean'),
    avg_severity=('severity', 'mean')
).reset_index().sort_values(by='avg_economic_impact_million_USD')
response_impact_correlation
```

```python
# ---------------------------------------------------
# Top 5 Deadliest Events Each Year (Window Function Equivalent)
# ---------------------------------------------------
# Filtering events with deaths > 0
deadliest_filtered = df_final[df_final['deaths'] > 0].copy()

# Simulating ROW_NUMBER() OVER (PARTITION BY year ORDER BY deaths DESC)
deadliest_filtered['rank_num'] = deadliest_filtered.groupby('year')['deaths'].rank(method='first', ascending=False).astype(int)

top_5_deadliest_per_year = deadliest_filtered[deadliest_filtered['rank_num'] <= 5][
    ['year', 'rank_num', 'event_id', 'country', 'event_type', 'deaths', 'economic_impact_million_usd']
].rename(columns={'economic_impact_million_usd': 'economic_impact_million_USD'}).sort_values(by=['year', 'rank_num'])
top_5_deadliest_per_year
```

```python
# ---------------------------------------------------
# International Aid Impact Analysis
# ---------------------------------------------------
df_final['aid_status'] = np.where(df_final['international_aid_million_usd'] > 0, 'Aid given', 'Without Aid')

aid_impact_analysis = df_final.groupby('aid_status').agg(
    event_count=('event_id', 'count'),
    avg_deaths=('deaths', 'mean'),
    avg_impact_million_USD=('economic_impact_million_usd', 'mean'),
    avg_damage_score=('infrastructure_damage_score', 'mean'),
    avg_response_time=('response_time_hours', 'mean')
).reset_index()
aid_impact_analysis
```

```python
# ---------------------------------------------------
# 1. Analysis of Response Efficiency (Speed vs Severity)
# ---------------------------------------------------
# Lower score means better efficiency (Fast response to a high severity event)
print("--- Response Efficiency Analysis by Continent ---")
avg_efficiency_by_geo = df_final.groupby('continent')['response_efficiency'].mean().reset_index()
# Sorting from most efficient (lowest score) to least efficient
avg_efficiency_by_geo = avg_efficiency_by_geo.sort_values(by='response_efficiency')
print(avg_efficiency_by_geo.to_string(index=False))
```

```python
print("\n--- Top 10 Most Efficiently Handled Events ---")
most_efficient_events = df_final[['event_id', 'country', 'event_type', 'severity', 'response_time_hours', 'response_efficiency']]\
    .sort_values(by='response_efficiency', ascending=True).head(10)
print(most_efficient_events.to_string(index=False))
```

```python
# ---------------------------------------------------
# 2. Analysis of Aid-to-Loss Ratio (Is financial aid enough?)
# ---------------------------------------------------
print("\n--- Average Aid-to-Loss Ratio by Country (Top 10) ---")
# Filtering only events that actually received aid to get an accurate ratio
aid_coverage_by_country = df_final[df_final['international_aid_million_usd'] > 0]\
    .groupby(['country', 'continent'])['aid_to_loss_ratio'].mean().reset_index()
aid_coverage_by_country = aid_coverage_by_country.sort_values(by='aid_to_loss_ratio', ascending=False).head(10)
print(aid_coverage_by_country.to_string(index=False))
```

```python
print("\n--- Correlation between Aid-to-Loss Ratio and Event Severity ---")
# Do severe events get proportionally more aid coverage?
aid_severity_corr = df_final[df_final['international_aid_million_usd'] > 0][['severity', 'aid_to_loss_ratio']].corr().iloc[0, 1]
print(f"Correlation Coefficient: {round(aid_severity_corr, 4)}")
```

```python
# ---------------------------------------------------
# 3. Analysis of Fatality Rate (Lethality of disasters)
# ---------------------------------------------------
print("\n--- Deadliest Disasters by Fatality Rate (Top 10) ---")
# Finding events where the percentage of deaths out of total casualties was the highest
highest_fatality_events = df_final[df_final['total_casualties'] > 0]\
    .sort_values(by='fatality_rate', ascending=False)\
    .head(10)[['event_id', 'country', 'event_type', 'deaths', 'total_casualties', 'fatality_rate']]
print(highest_fatality_events.to_string(index=False))
```

```python
print("\n--- Average Fatality Rate by Event Family ---")
fatality_by_family = df_final[df_final['total_casualties'] > 0]\
    .groupby('event_family')['fatality_rate'].mean().reset_index()
fatality_by_family = fatality_by_family.sort_values(by='fatality_rate', ascending=False)
print(fatality_by_family.to_string(index=False))
```

```python
# ---------------------------------------------------
# 1. Aid vs Response Efficiency Matrix
# ---------------------------------------------------
print("--- International Aid vs Response Efficiency (By Continent) ---")
# Lower response_efficiency + Higher aid_to_loss_ratio = Well funded and fast handled
aid_vs_speed = df_final.groupby('continent').agg(
    avg_aid_ratio=('aid_to_loss_ratio', 'mean'),
    avg_response_efficiency=('response_efficiency', 'mean'),
    total_events=('event_id', 'count')
).reset_index().sort_values(by='avg_response_efficiency')

print(aid_vs_speed.to_string(index=False))
```

```python
# ---------------------------------------------------
# 2. Climate Resilience Index (Top 10 Most Resilient Countries)
# ---------------------------------------------------
print("\n--- Top 10 Most Resilient Countries ---")
# Definition: High average severity, but low fatality rate and low infrastructure damage
country_resilience = df_final.groupby(['country', 'continent']).agg(
    event_count=('event_id', 'count'),
    avg_severity=('severity', 'mean'),
    avg_fatality_rate=('fatality_rate', 'mean'),
    avg_infra_damage=('infrastructure_damage_score', 'mean')
).reset_index()

# Filter out countries with very few events to avoid bias (e.g., keeping countries with > 2 events)
min_events = country_resilience['event_count'].median()
resilient_countries = country_resilience[country_resilience['event_count'] >= min_events].copy()

# Resilience Score Formula: Severity divided by (Fatality Rate + Damage Score + 1)
# Higher score = More resilient (Handles severe events with minimal loss)
resilient_countries['resilience_score'] = resilient_countries['avg_severity'] / (
    resilient_countries['avg_fatality_rate'] + resilient_countries['avg_infra_damage'] + 1
)

top_resilient = resilient_countries.sort_values(by='resilience_score', ascending=False).head(10)
print(top_resilient[['country', 'continent', 'event_count', 'avg_severity', 'resilience_score']].to_string(index=False))
```

```python
# ---------------------------------------------------
# 3. Extreme Events Profile (Events with Catastrophic Impact)
# ---------------------------------------------------
print("\n--- Extreme Events Profile Analysis ---")
# Defining extreme events: Severity > 7 OR Economic Impact > 100 Million USD
extreme_events = df_final[(df_final['severity'] > 7) | (df_final['economic_impact_million_usd'] > 100)].copy()

print(f"Total Extreme Events Identified: {len(extreme_events)}")

extreme_analysis = extreme_events.groupby('event_family').agg(
    extreme_event_count=('event_id', 'count'),
    total_extreme_deaths=('deaths', 'sum'),
    avg_duration_days=('duration_days', 'mean'),
    avg_fatality_rate=('fatality_rate', 'mean')
).reset_index().sort_values(by='extreme_event_count', ascending=False)

print(extreme_analysis.to_string(index=False))
```

```python
#  DIAGNOSTIC & EXPLORATORY DATA ANALYSIS (DIAGNOSTIC)

# ---------------------------------------------------
# 1. Geographical Aid Gap Analysis (Testing Funding Fairness)
# ---------------------------------------------------
print("--- Geographical Aid Gap Analysis (High Severity Events > 6) ---")
# Filtering only severe disasters to see how aid is distributed across continents
severe_aid_distribution = df_final[df_final['severity'] > 6].groupby('continent').agg(
    severe_event_count=('event_id', 'count'),
    avg_aid_million_USD=('international_aid_million_usd', 'mean'),
    avg_aid_to_loss_ratio=('aid_to_loss_ratio', 'mean'),
    total_deaths=('deaths', 'sum')
).reset_index().sort_values(by='avg_aid_to_loss_ratio', ascending=False)

print(severe_aid_distribution.to_string(index=False))
```

```python
# ---------------------------------------------------
# 2. Temporal Peaks & Cumulative Impact Analysis
# ---------------------------------------------------
print("\n--- Top 5 Worst Timeline Peaks (Year + Month Cumulative Impact) ---")
# Finding months in history where the cumulative pressure on humanity was at its peak
timeline_peaks = df_final.groupby(['year', 'month_name']).agg(
    events_in_month=('event_id', 'count'),
    cumulative_deaths=('deaths', 'sum'),
    cumulative_impact_million_USD=('economic_impact_million_usd', 'sum'),
    avg_response_hours=('response_time_hours', 'mean')
).reset_index().sort_values(by='cumulative_impact_million_USD', ascending=False).head(5)

print(timeline_peaks.to_string(index=False))
```

```python
# ---------------------------------------------------
# 3. Weekend vs. Weekday Emergency Response Test
# ---------------------------------------------------
print("\n--- Emergency Response Test: Weekends vs. Weekdays ---")
# Diagnostic Test: Does the response time drop during weekends when offices are closed?
weekend_vs_weekday = df_final.groupby('is_weekend').agg(
    event_count=('event_id', 'count'),
    avg_response_time_hours=('response_time_hours', 'mean'),
    avg_response_efficiency=('response_efficiency', 'mean'),
    avg_infrastructure_damage=('infrastructure_damage_score', 'mean')
).reset_index()

# Renaming for better readability in output
weekend_vs_weekday['day_type'] = np.where(weekend_vs_weekday['is_weekend'], 'Weekend', 'Weekday')
print(weekend_vs_weekday[['day_type', 'event_count', 'avg_response_time_hours', 'avg_response_efficiency']].to_string(index=False))
```
